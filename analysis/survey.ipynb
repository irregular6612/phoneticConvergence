```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
```

```python
filePath = "../../../../Desktop/results/survey_results_0429.xlsx"
```

```python
## 폰트 확인
import matplotlib.font_manager as fm
for f in fm.fontManager.ttflist:
    if 'Gothic' in f.name or 'Apple' in f.name or 'Nanum' in f.name:
        print(f.name)
```

```python
# 나눔고딕 설치 후(필요시), 아래처럼 사용
plt.rcParams['font.family'] = 'Nanum Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
data_df = pd.read_excel(filePath)[6:]
data_df = data_df.reset_index()
data_df.drop(columns=['index'], inplace=True)
data_df.columns = ['timestamp', 'participant_no','Who', 'AI-Like', 'Human-Like', 'List', 'etc1', 'etc2']
data_df

sns.countplot(data=data_df, x="Who")
```

```python
# List 1: AI라고 속인 경우, List 2: 사람이라고 밝힌 경우,
sns.countplot(data=data_df, x="AI-Like", hue="List")
```
