```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
controlFilePath = "../../../../Documents/WorkSpace/phoneticConvergence/audio-sample/short-version-phonetic/control_data-analysis.xlsx"
```

```python
control_data_df = pd.read_excel(controlFilePath, sheet_name="Stage")
control_data_df
```

```python
vowel_set = ['ㅏ', 'ㅐ', 'ㅣ', 'ㅗ', 'ㅜ']
```

```python
vowel_df = pd.DataFrame()
vowel_df['vowel'] = pd.concat([control_data_df['first_vowel'], control_data_df['second_vowel']], axis=0)
vowel_df['f1'] = pd.concat([control_data_df['first_f1'], control_data_df['second_f1']], axis=0)
vowel_df['f2'] = pd.concat([control_data_df['first_f2'], control_data_df['second_f2']], axis=0)
vowel_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 한글 폰트 설정
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# 모음별로 그룹화
grouped = vowel_df.groupby('vowel')

# 모음 목록 가져오기
vowels = vowel_df['vowel'].unique()

# 평균값 표시를 위한 스타일 설정
meanprops = dict(marker='D', markeredgecolor='black', markerfacecolor='red')

# 각 모음별로 박스플롯 그리기
for vowel in vowels:
    # 해당 모음 데이터 가져오기
    group = grouped.get_group(vowel)
    
    # 2개의 서브플롯이 있는 figure 생성
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # f1 박스플롯
    sns.boxplot(y=group['f1'], ax=axes[0], color='skyblue', showmeans=True, meanprops=meanprops)
    axes[0].set_title(f'모음 "{vowel}"의 F1 분포')
    axes[0].set_ylabel('F1 (Hz)')
    
    # F1 평균값과 중앙값 표시
    f1_mean = group['f1'].mean()
    f1_median = group['f1'].median()
    axes[0].text(0.5, 0.95, f'평균: {f1_mean:.1f} Hz\n중앙값: {f1_median:.1f} Hz',
                transform=axes[0].transAxes, ha='center', va='top',
                bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))
    
    # f2 박스플롯
    sns.boxplot(y=group['f2'], ax=axes[1], color='lightgreen', showmeans=True, meanprops=meanprops)
    axes[1].set_title(f'모음 "{vowel}"의 F2 분포')
    axes[1].set_ylabel('F2 (Hz)')
    
    # F2 평균값과 중앙값 표시
    f2_mean = group['f2'].mean()
    f2_median = group['f2'].median()
    axes[1].text(0.5, 0.95, f'평균: {f2_mean:.1f} Hz\n중앙값: {f2_median:.1f} Hz',
                transform=axes[1].transAxes, ha='center', va='top',
                bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))
    
    # 전체 제목 설정
    plt.suptitle(f'모음 "{vowel}"의 F1 및 F2 분포', fontsize=16)
    
    # 레이아웃 조정
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    
    # 그래프 표시
    plt.show()
```

```python
# 한글 폰트 설정
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# 하나의 figure에 두 개의 서브플롯 생성
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# F1 박스플롯 (왼쪽)
sns.boxplot(x='vowel', y='f1', data=vowel_df, ax=axes[0], palette='Set3')
axes[0].set_title('모음별 F1 분포')
axes[0].set_xlabel('모음')
axes[0].set_ylabel('F1 (Hz)')

# F2 박스플롯 (오른쪽)
sns.boxplot(x='vowel', y='f2', data=vowel_df, ax=axes[1], palette='Set3')
axes[1].set_title('모음별 F2 분포')
axes[1].set_xlabel('모음')
axes[1].set_ylabel('F2 (Hz)')

# 전체 제목 설정
plt.suptitle('모음별 F1 및 F2 분포 비교', fontsize=16)

# 레이아웃 조정
plt.tight_layout()
plt.subplots_adjust(top=0.85)

# 그래프 표시
plt.show()
```
